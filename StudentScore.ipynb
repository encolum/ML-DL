```python
x = df.drop("writing score", axis = 1)
y = df["writing score"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
#SimpleImputer: điền null 
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy= 'median')
x_train[["math score", "reading score"]] = imputer.fit_transform(x_train[["math score", "reading score"]])
```

```python
from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()
# x_train[["math score", "reading score"]] =scaler.fit_transform(x_train[["math score", "reading score"]])
```

```python
#Pipeline
from sklearn.pipeline import Pipeline
num_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])
result = num_transformer.fit_transform(x_train[['math score', 'reading score']])
for i, j in zip(x_train[["math score", "reading score"]].values, result):
    print("Before {}. After {}".format(i,j))
```

```python
#NominalEncoder
from sklearn.preprocessing import OneHotEncoder
nom_transform = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy= "most_frequent")),
    ("encoder" , OneHotEncoder(sparse_output= False))
])
result2 = nom_transform.fit_transform(x_train[["race/ethnicity"]])
for i, j in zip(x_train[["race/ethnicity"]].values,result2 ):
    print(f"Before {i}. After {j}")
#(0,3) 1.0 là các phần từ khác = 0 và phần tử thứ 3 ( tính từ index = 0) là 1.0
#Ví dụ (0,3) 1.0 là vector 0 0 0 1 0 

```

```python
#Boolean variables: Use ordinal encoder 
education_values = ['some high school', 'high school' , 'some college', "associate's degree", "bachelor's degree", "master's degree"]
gender_values = ["male" , "female"]
lunch_values = x_train["lunch"].unique()
test_values = x_train["test preparation course"].unique()
ord_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OrdinalEncoder(categories= [education_values, gender_values, lunch_values, test_values]))
])
result1 = ord_transformer.fit_transform(x_train[["parental level of education", "gender", "lunch", "test preparation course"]])
```

```python
#ColumnTransformer
from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer(transformers=[
    ("num_feature", num_transformer, ["reading score", "math score"]),
    ("ord_feature", ord_transformer, ["parental level of education", "gender", "lunch", "test preparation course"]),
    ("nom_feature", nom_transform, ["race/ethnicity"]),
])
```

```python
#Test thử mô  hình Random Forest 
from sklearn.ensemble import RandomForestRegressor
reg2 = Pipeline(steps =[
    ("preprocessor", preprocessor),
    ("model" , RandomForestRegressor())
])
reg2.fit(x_train, y_train)
y_predict2 = reg2.predict(x_test)
# for i, j in zip(y_predict, y_test):
#     print(f"Predicted value {i}. Actual value: {j}")
print(f"MAE: {mean_absolute_error(y_test, y_predict2)}")
print(f"MsE: {mean_squared_error(y_test, y_predict2)}")
print(f"R2: {r2_score(y_test, y_predict2)}")



#GridSearch
from sklearn.model_selection import GridSearchCV
params ={
    "preprocessor__num_feature__imputer__strategy":["median", "mean"],
    "model__n_estimators": [100,200,300],
    "model__criterion" : ["poisson", "absolute_error", "squared_error"],
    # "model__max_depth": [None,2 ,5 ]

}
#Có randomizesearch để search ví dụ như 20 trường hợp 
grid_search = GridSearchCV(estimator=reg2, param_grid=params, cv=5, scoring= "r2", verbose=2, n_jobs= 1) #n_job là số nhân
#cv: cross-validation,  verbose: số càng cao in càng nhiều
grid_search.fit(x_train,y_train)
y_predict3 = grid_search.predict(x_test)
```

```python
print(grid_search.best_estimator_)
print(grid_search.best_score_)
print(grid_search.best_params_)
print(f"MAE: {mean_absolute_error(y_test, y_predict3)}")
print(f"MsE: {mean_squared_error(y_test, y_predict3)}")
print(f"R2: {r2_score(y_test, y_predict3)}")
```
