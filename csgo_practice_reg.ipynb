```python
import pandas as pd
from ydata_profiling import ProfileReport
import numpy as np
```

```python
df = pd.read_excel("csgo.xlsx")
df.head()
```

```python
y = df['points']
x = df.drop(['day', 'month', 'year', 'date', 'match_time_s', 'team_a_rounds', 'team_b_rounds', 'wait_time_s', 'result', 'points'], axis=1)
print(f"X: \n{x.head()}")
print(f"Y:\n {y.head()}")
```

```python
from sklearn.model_selection import train_test_split
#SimpleImputer: điền null 
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
from lazypredict.Supervised import LazyRegressor
clf = LazyRegressor(verbose=0,ignore_warnings=True, custom_metric=None)
models,predictions = clf.fit(x_train, x_test, y_train, y_test)
models
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
```

```python
#Pipeline
from sklearn.pipeline import Pipeline
num_transform = Pipeline(steps =[
    ('imputer', SimpleImputer(strategy= 'mean')),
    ('scaler', StandardScaler())
])

nom_transform = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy= 'most_frequent')),
    ('encoder', OneHotEncoder(sparse_output=False))
])
num_features = []
num_features_value = []
cols = x_train.columns
for i in cols:
    if x_train[i].dtypes == 'int64':
        num_features.append(i)
        num_features_value.append(x_train[i].unique())
```

```python
from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer(transformers=[
    ('num_features', num_transform, num_features),
    ('nom_transform', nom_transform, ['map'])
    
])
```

```python
#Train model
reg = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', LinearRegression())
])
reg.fit(x_train,y_train)
```

```python
y_predict = reg.predict(x_test)
for i, j in zip(y_predict, y_test):
    print(f'Predict: {i} ---> Actual: {j}')
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print(f"R2: {r2_score(y_test, y_predict)}")
print(f"MSE: {mean_squared_error(y_test, y_predict)}")
```
