```python
target = 'Outcome'
x = data.drop(target, axis = 1)
y = data[target]
```

```python
x
```

```python
y
```

```python
#Data split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
# x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.25)
```

```python
x_train
```

```python
print(y_train.value_counts()) #in ra gia so luong gia tri cua y
print(y_test.value_counts())
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
#fit: scaler.fit(x_train) 
#transform: x_train = scaler.transfrom(x_train)
#fit_transform : x_train = scaler.fit_transform(x_train)
#x_test = scaler.transform(x_test)
#Bộ train thì dùng fit_transform, bộ test thì dùng transform
#Không được dùng fit với bộ test (data leakage)
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
#Correlation > 0,7 or < -0.7 thì mới dùng mô hình tuyến tính(Logistic Regression)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
#Train model
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
print(y_predict)
print()
print(y_test.values)
```

```python
for i, j in zip(y_predict,y_test.values):
    print(f'Prediction: {i}. Actual value: {j}')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```

```python
#Vi outlier nên hệ số tương quan (correlation ) bị sai lệch
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model1 = LogisticRegression()
model1.fit(x_train,y_train)
y_predict1 = model1.predict(x_test)
print(classification_report(y_test, y_predict1))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model2 = RandomForestClassifier(n_estimators=200,random_state=100)
model2.fit(x_train,y_train)
y_predict2 = model2.predict(x_test)
print(classification_report(y_test, y_predict2))
```

```python
#K-fold cross validation
```

```python
#GridSearch
from sklearn.model_selection import GridSearchCV
params ={
    "n_estimators": [100,200,300],
    "criterion" : ["gini", "entropy", "log_loss"]

}
grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=100), param_grid=params, cv=4, scoring= "recall", verbose=2)
grid_search.fit(x_train,y_train)
print(grid_search.best_estimator_)
print(grid_search.best_score_)
print(grid_search.best_params_)
```

```python
y_predict3 = grid_search.predict(x_test)
print(classification_report(y_test, y_predict3))
```
